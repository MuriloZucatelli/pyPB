# Teste de otimização

```python
from scipy.optimize import minimize, differential_evolution, rosen, rosen_der
from numpy import array, set_printoptions
set_printoptions(precision=3)
def counted(f):
    def wrapped(*args, **kwargs):
        wrapped.calls += 1
        print(*args[1], f(*args))
        return f(*args, **kwargs)
    wrapped.calls = 0
    return wrapped

class newrosen:
    @counted
    def n_rosen(self, *args, **kwargs):
        return rosen(*args, **kwargs)
```

```python

x0 = array([1.3, 0.7, 0.8, 1.9, 1.2])
results = []
res = dict()
newr = newrosen()
Copt = minimize(newr.n_rosen,
    x0,
    method="L-BFGS-B", #jac = True
    bounds=[(0, None)] * 5,
    options={
        "disp": True,
        "ftol": 1e-8,  # default: 2.2e-09
        "gtol": 1e-6,  # 1e-8 default is 1e-5
        "maxls": 20,  # Max line search steps (per iteration). Default is 20.
        "maxiter": 20,
        # "eps": 0.1, # eps: default 1e-8
    },
)  # default: 15000
# GLOBAL OPTIMIZATION
# b = [(0, 1), (0, 1), (0, 10), (0, 10)]
# Copt = differential_evolution(lambda c: error_function(c, e),
# bounds=b, maxiter=10, popsize=1)
res["best_fit"] = Copt.x
results.append(res)
Copt
```

```python
class breakup_models:
    def __init__(
        self, name, C
    ) -> None:
        self.C = C
        self.models = {}

        if name == 'coulaloglou':
            self.gamma = self.coulaloglou
        else:
            self.gamma = None
    
    def coulaloglou(self,v):
        C1 = self.C[0]
        return 1+v*C1**2
```
